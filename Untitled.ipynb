```python
# h = [170,180,175,160]
# f = [260,270,265,255]

h1 = 170
f1 = 260

# 신발 = 키 * a + b

a = tf.Variable(0.1)
b = tf.Variable(0.2)

def loss_func(): # 손실함수
    expect = h1 * a + b
    return tf.square(f1 - expect) # (실제값 - 예측값)^2

opt = tf.keras.optimizers.Adam(learning_rate=0.1) # 경사하강법

for i in range(300):
    opt.minimize(loss_func,var_list=[a,b]) # 손실함수(loss_func), 경사하강법으로 업데이트할 weight Variable 목록
    print(a.numpy(),b.numpy())

```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('gpascore.csv')
# print(data.isnull().sum())
data = data.dropna()

y = data['admit'].values
x = []

for i, rows in data.iterrows():
    x.append([rows['gre'], rows['gpa'], rows['rank']])


model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64,activation="tanh"), # 노드 갯수
    tf.keras.layers.Dense(128,activation="tanh"),
    tf.keras.layers.Dense(1,activation="sigmoid"), 
    # 마지막 레이어- 예측결과니까 0-1 확률 하려면 sigmoid 쓰면 됨
]) # 딥러닝 모델

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) #adam 맛좋음.  예측할 때 - loss는 binary_crossentropy(결과가 0과 1사이의 분류/확률문제)

# x데이터 : [[380,3.21,3],[660,3.67,3], [], [], ...]
# y데이터 : [[0], [1], [1], ...]

model.fit(np.array(x),np.array(y),epochs=500) # 학습을 해 준다. 인자에는 x데이터(학습데이터), y데이터(실제정답), epochs=1

# 예측
predict = model.predict([[750,3.70,3],[400,2.2,1]]) # model.predict(테스트해볼 x데이터)
print(predict)
```
